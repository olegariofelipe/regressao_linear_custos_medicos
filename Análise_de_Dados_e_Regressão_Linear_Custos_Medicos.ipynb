###__1.2. Lidando com outliers:__

```python
# Função para identificar outliers:
def outliers(df, col):
  q1 = df[col].quantile(.25)
  q3 = df[col].quantile(.75)
  iqr = q3 - q1
  limite_inferior = q1 - 1.5 * iqr
  limite_superior = q3 + 1.5 * iqr
  outliers = df[(df[col] < limite_inferior) | (df[col] > limite_superior)].index
  return outliers
```

```python
# Obtendo os outliers da coluna charges:
outliers_charges = dados_unicos.iloc[outliers(dados_unicos, 'charges')]
display(outliers_charges.describe().round(2).T, outliers_charges.shape)
```

```python
# Valores unicos de cada outlier - Charges:
np.sort(outliers_charges['charges'].unique())
```

```python
# Verificando proporção de outliers em relação a base de dados:
prop = (outliers_charges.shape[0] / dados_unicos.shape[0]) * 100
print(f'Proporção = {prop:.2f}%')
```

```python
# Verificando os valores mínimo e máximo para a idade:
print(dados_unicos['age'].min(), dados_unicos['age'].max())
```

```python
# Organizandos valores para realizar classificação dos dados:
classes = [dados_unicos['age'].min(), 21, 26, 41, dados_unicos['age'].max()]
labels = ['Adolescencia', 'Pós-Adolecência', 'Adultidade', 'Meia-Idade']

# Data frame que receberá as idade classificadas:
class_age = pd.DataFrame()

# Inserindo as colunas necessárias:
class_age['classes'] = pd.cut(x = dados_unicos['age'], bins = classes, labels = labels, include_lowest = True)
class_age['age'] = dados_unicos['age']
class_age['charges'] = dados_unicos['charges']
class_age.head()
```

Isso faz sentido, pois podemos assumir que com o avanço da idade, os custos tendem a aumentar pois aumentam os problemas de saúde, porém isso requer uma análise mais aprofundada.

__2.1.1. Podemos inferir  que os valores de custos do sexo masculino também são maiores para toda a população.__

```python
# Realizando teste de normalidade:
from scipy.stats import normaltest

estatistica, p_valor = normaltest(dados_unicos['charges'])
print(f'p-valor: {p_valor}')
```

Assumindo uma significância de 5%, podemos rejeitar a hipótese nula de que os dados são provenintes de uma distribuição normal.


__Teste de Hipótese: Os gastos são maiores para pessoas do sexo masculino ?__

* $u_m$: média de custos dos homens.
* $u_f$: média de custos das mulheres.

$H_0$: $u_m \leq u_f$

$H_1$: $u_m > u_f$

__Como os dados não se distribuem de forma normal, será necessário utilizar um teste não paramétrico, que nesse caso será o teste de Mann-Whitney para comparação das duas médias__.

```python
# Importação de módulo necessário
from scipy.stats import mannwhitneyu as U

# Amostras:
homens = dados_unicos.query('sex == "male"')['charges']
mulheres = dados_unicos.query('sex == "female"')['charges']

# Obtenção dos resultados
u, p = U(homens, mulheres, alternative='greater')

# Imprimindo na tela:
resultado = 'Rejeitar H0' if p < 0.05 else 'Aceitar H0'
print(resultado)
```

__Conclusão:__ _Com um nível de significância de 5%, não existem evidências o suficiente para rejeitar a hipótese de que média de gastos das pessoas de sexo masculino é igual a média de gastos das pessoas do sexo femino. Ou seja, não podemos concluir que existe diferença significativa na média de gastos._

```python
# Custo médio por região:

# Agrupando custos por região
charges_region = dados_unicos.groupby('region')['charges'].mean().round(2).sort_values(ascending=False).reset_index()

# Criando e configurando gráfico:
ax = sns.barplot(data = charges_region, x = 'charges', y = 'region', orient = 'h', hue='region')
ax.set_title('Custo médio por região', fontweight='bold', fontsize=16, loc='left', y=1.05)
ax.set_ylabel('Região', fontsize=12, fontweight='bold')
ax.set_xlabel('Custo médio', fontsize=12, fontweight='bold')

# Remoção das linhas de caixa do gráfico:
sns.despine()

# Adicionando labels de cada barra do gráfico:
for container in ax.containers:
  ax.bar_label(container, fmt = '%2g', fontweight='bold')

# Mostrando gráfico
plt.show()
```

```python
# Realizando o cruzamento de informações:
mean_charges_smoker = pd.crosstab(
                                  index = class_age['smoker'],
                                  columns = class_age['classes'],
                                  aggfunc = 'mean',
                                  values = class_age['charges']
                                  ).round(2)

mean_charges_smoker
```

#### __Conclusão:__

O tabagismo afeta de forma significativa nos custos. É possível verificar que em todas as faixas etárias existe uma grande discrepancia de custos entre os fumantes e não fumantes.

# **4. Costrução de Modelo de Regressão Linear com Scikit-Learn**

__Estimando modelo com dados de treino:__

```python
# Importação de pacotes necessários:
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
# Instanciando modelo:
modelo_skl = LinearRegression()
```

```python
# Divisao dos dados:
X = dados_regressao.drop(['charges','sex'], axis=1)
y = dados_regressao['charges']

# Disvisão das bases de dados:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, random_state = 42)
display(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Treinando modelo:
modelo_skl.fit(X_train, y_train)
```

```python
# Obtendo R² para os dados de treino:
R2 = modelo_skl.score(X_train, y_train).round(2)
print(f'R² = {R2:.2f}')
```

```python
# Realizando previsões para dados de treino:
y_previsto = modelo_skl.predict(X_test)
```

```python
# Obtendo R² para os dados de teste:
print(f'R² = {r2_score(y_test, y_previsto):.2f}')
```

O modelo consegue descrever em 77% da variabilidade dos dados de teste.

__Calculando outras métricas para o modelo:__

```python
# Calculo Erro Quadratico Médio:
EQM = metrics.mean_squared_error(y_test, y_previsto).round(2)

# Calculo da Raiz do Erro Quadratico Médio:
SEQM = np.sqrt(EQM).round(2)

pd.DataFrame({'valores':[EQM,SEQM,R2]}, index=['EQM','SEQM','R²'])
```

### Gerando previsões para os dados de treino:


```python
# Realizando predições com dados de treino:
y_previsto_treino = modelo_skl.predict(X_train)
```

```python
# Visualizando os resultados:

# Criando e configurando gráfico:
plt.figure(figsize=(10,6))
plt.suptitle('Previsão vs Real', fontweight='bold', fontsize=16)
ax = sns.scatterplot(x=y_train, y=y_previsto_treino, label='Treino')
ax = sns.scatterplot(x=y_test, y=y_previsto, label='Teste')
ax.set_xlabel('Real', fontweight='bold')
ax.set_ylabel('Previsão', fontweight='bold')

# Mostrando gráfico:
plt.show()
```

#### __Conclusão:__ Por mais que o modelo consiga explicar 74% a variação de custos, ele ainda apresenta um erro médio quadratico bem elevado, analisando o gráfico acima, vemos que o modelo não apresenta grande variação em relação as previsões de treino e de teste.
